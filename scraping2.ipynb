```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


with open('test.html') as file:
    html = file.read()



soup = BeautifulSoup(html, 'html.parser')

# Find all elements with the "name" and "quantity" classes
name_elements = soup.find_all(class_="_1BeXdImwm55uhlaM6taNA7 _1IaHHFR-s586srJTVU0l7N _2xXivNl-PoHSBMShYslSQd")
quantity_elements = soup.find_all(class_="lUfd37k7gZ4cB04c6Ndb9")
price_elements = soup.find_all(class_="_2UgRR0oHP74fCvcydTbd3B")

# List to store the extracted data
data = []

# Extract data from the elements
for i in range(len(name_elements)):
    name_value = name_elements[i].find('span').get_text().replace("\n","").replace("\r","")
    clean_title = ' '.join(name_value.split())
    spans = quantity_elements[i].find_all('span')
    QYT = ""
    for span in spans:
        QYT += span.text +" "
    
    clean_quantity = ' '.join(QYT.split())

    price_value = price_elements[i].get_text()
    print(clean_quantity)


    data.append([clean_title, clean_quantity, price_value])

# Create a DataFrame with the extracted data
df = pd.DataFrame(data, columns=["Name", "Quantity","Price"])

# Save the DataFrame to an Excel file
df.to_excel("data.xlsx", index=False)
```
